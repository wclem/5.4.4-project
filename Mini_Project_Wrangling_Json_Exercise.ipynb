## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
json.load((open('data/world_bank_projects.json')))
wb_df = pd.read_json('data/world_bank_projects.json')
#wb_df
```

```python
# Q1 Top Ten countries with most projects
top_countries = wb_df.groupby('countryname').size()
top10 = top_countries.sort_values(ascending=False)[:11] #Showing top 11 because Africa is not a country 
top10
```

```python
#Q2 Top 10 Themes
themes = wb_df['mjtheme_namecode']
theme_names = []
for row in themes:
   for name_item in row:
       theme_names.append(name_item['name'])
themes_df = pd.DataFrame({'theme names': theme_names})
theme_count = themes_df['theme names'].value_counts(sort=True)
theme_count
# Appers that 122 projects had no theme names specified
# Also, some themes are missing, even though a code is specified.
```

```python
# Q3: Fill in missing theme names
# Start by counting all the codes, to identify what codes are used.
codes = []
for row in themes:
   for name_item in row:
    if name_item['name'] == '':
       codes.append(name_item['code'])
codes_df = pd.DataFrame({'codes': codes})
codes_count = codes_df['codes'].value_counts(sort=True)
codes_count
# Codes are numbered 1 - 11
```

```python
#Used this code to investigate which codes match which code themes
code_theme_df = wb_df[['mjtheme', 'mjthemecode']]
for i in range(499):
    codes = code_theme_df.loc[i, 'mjthemecode']
    if '3' in codes:
        print(code_theme_df.loc[i, ['mjtheme', 'mjthemecode']])
```

```python
print(type(code_theme_df['mjtheme']))
```

```python
#Create a dict of code nums and themes
themecodes= {1: 'Economic management', 2: 'Public sector governance', 3: 'Rule of law',
            4: 'Financial and private sector development', 5: 'Trade and integration',
            6: 'Social protection and risk management', 7: 'Social dev/gender/inclusion',
            8: 'Human development', 9: 'Urban development', 10: 'Rural development',
            11: 'Environment and natural resources management'}
themecodes
```

```python
# Function to create list of themes from code numbers, each project usually has multiple codes
def decode_themes(code_list):
    tcodes = code_list.split(',')
    themes = []
    for i in tcodes:
        themes.append(themecodes[int(i)])
    return themes

print(decode_themes('1,2,3'))
```

```python
# Update the project list with complete list of themes
corrected = code_theme_df['mjthemecode'].apply(decode_themes)
new_code_theme_df = code_theme_df.merge(corrected.rename('corrected themes'), left_index=True, right_index=True)
new_code_theme_df
```

```python
# Find Top 10 Themes using updated dataframe with complete theme lists
corr_themes = new_code_theme_df['corrected themes']
theme_names = []
for row in corr_themes:
   for name_item in row:
      theme_names.append(name_item)
themes_df = pd.DataFrame({'theme names': theme_names})
new_theme_count = themes_df['theme names'].value_counts(sort=True)
print('Original theme count\n')
print(theme_count)
print('\nCorrected theme count\n')
print(new_theme_count)
```
